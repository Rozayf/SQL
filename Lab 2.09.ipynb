```python
import pandas as pd
import numpy as np
import pymysql 
from sqlalchemy import create_engine
from getpass import getpass
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
data_actor = pd.read_sql_query('SELECT * FROM sakila.actor', engine)
data_actor
```

```python
disctinctactor_count = pd.read_sql_query('SELECT COUNT(DISTINCT last_name) AS actorcount FROM actor', engine)
```

```python
disctinctactor_count
```

```python
weekday_weekend = pd.read_sql_query('''SELECT *,
CASE
WHEN WEEKDAY(rental_date) >= 6 then 'weekend'
WHEN WEEKDAY(rental_date) < 6 then 'weekday'
END AS 'day_type', WEEKDAY(rental_date) 'weekday'
FROM rental;''', engine)
```

```python
weekday_weekend
```

```python
data_film = pd.read_sql_query('SELECT * FROM sakila.film', engine)
data_film
```

```python
similartitle = pd.read_sql_query("SELECT title FROM film WHERE LEFT (title,10) = 'ARMAGEDDON' or RIGHT (title,10) = 'ARMAGEDDON';", engine)
```

```python
similartitle
```

```python
longest_films = pd.read_sql_query('''SELECT title, length from film
ORDER BY length DESC
LIMIT 10;''',engine)
```

```python
longest_films
```

```python
behindthescenes = pd.read_sql_query('''SELECT COUNT(title) FROM film
WHERE special_features LIKE '%%Behind the Scenes%%';''',engine)
```

```python
behindthescenes
```

```python
rating_duration = pd.read_sql_query('''SELECT COUNT(title) as title_count, concat(floor(AVG(length)/60),'h', round(AVG(length)%%60,0),'m')AS average_duration, rating FROM film
GROUP BY rating;''', engine)
```

```python
rating_duration
```
